```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def compare_images(image_path1, image_path2):
    # 1. 이미지 로드
    img1 = cv2.imread(image_path1)
    img2 = cv2.imread(image_path2)

    if img1 is None or img2 is None:
        print("이미지를 찾을 수 없습니다.")
        return

    # 2. 크기 비교 (크기가 다르면 비교 불가)
    if img1.shape != img2.shape:
        print(f"이미지 크기가 다릅니다. img1: {img1.shape}, img2: {img2.shape}")
        return

    # -----------------------------------------------------------
    # Method 1: 단순 픽셀 일치 여부 (완전 동일한 파일인가?)
    # -----------------------------------------------------------
    is_identical = np.array_equal(img1, img2)
    print(f"1. 완전 동일 여부 (Exact Match): {is_identical}")

    # -----------------------------------------------------------
    # Method 2: MSE (Mean Squared Error) & PSNR
    # -----------------------------------------------------------
    # MSE: (A - B)^2 의 평균
    # float로 변환하여 오버플로우 방지
    mse = np.mean((img1.astype("float") - img2.astype("float")) ** 2)
    
    # PSNR: OpenCV 내장 함수 사용 (또는 수식으로 직접 구현 가능)
    if mse == 0:
        psnr = float('inf') # 차이가 없으면 무한대
    else:
        psnr = cv2.PSNR(img1, img2)
        
    print(f"2. MSE (작을수록 좋음): {mse:.4f}")
    print(f"3. PSNR (클수록 좋음): {psnr:.4f} dB")

    # -----------------------------------------------------------
    # Method 3: SSIM (구조적 유사도) - 사람의 눈과 가장 비슷
    # -----------------------------------------------------------
    # channel_axis=2: 컬러 이미지(H, W, C)일 경우 채널 축 지정
    score, diff = ssim(img1, img2, full=True, channel_axis=2)
    print(f"4. SSIM (1.0에 가까울수록 좋음): {score:.4f}")

    return mse, psnr, score

# 실행 예시
compare_images('/home/qkrwnstj/Style-Transfer/ours_2/yr_1203_2alpha_0.5_2_0/content_005_style_013_style_014_ratio_0.png', '/home/qkrwnstj/Style-Transfer/ours_2/yr_1203_2alpha_0.5_2_0/content_005_style_013_style_014_ratio_0.5.png')
```
